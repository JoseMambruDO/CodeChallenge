```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],

                    'value': [1, 2, 3, 5]})
```

```python
df1
```

```python
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],

                    'value': [5, 6, 7, 8]})
```

```python
df2
```

```python
pd.merge(df1,df2,on='value') #using value field
```

```python
df1.merge(df2, left_on='lkey', right_on='rkey') #using lkey and rkey, adding a suffix _x(right) or _y(left)
```

```python
df1.merge(df2, left_on='lkey', right_on='rkey', suffixes=('_al', '_br')) #with new suffix
```

```python
pd.merge(df1,df2,on='value', how='right') #right join
```

```python
pd.merge(df1,df2,on='value', how='left') #left join
```

```python
pd.merge(df1,df2,on='value', how='outer') #outer join
```

```python
df3 = pd.DataFrame({'key':['AA','AB','AC','AC','AB'],'values': range(5)})
df3
```

```python
df4 = pd.DataFrame({'cat':['100','200']},index=['AC','AB'])
df4
```

```python
pd.merge(df3,df4, left_on='key',right_index=True) #using right with index
```
