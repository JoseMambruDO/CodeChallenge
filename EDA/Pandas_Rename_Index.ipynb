```python
import numpy as np
import pandas as pd
```

```python
df1 = pd.DataFrame(np.arange(16).reshape(4,4),
                   index=['DN','SDN','SDO','SDE'],
                  columns=['idx','lat','lon','cap'])
df1
```

```python
df1.index = df1.index.map(str.lower)
df1
```

```python
df1.columns = df1.columns.map('c_{}'.format)
df1
```

```python
#using rename
df1.rename(index=str.title, columns=str.upper)
```

```python
#rename using dict
df1.rename(index={'dn':'DIST_NAC'}, columns={'c_idx':'INDEX'},inplace=True)
df1
```
